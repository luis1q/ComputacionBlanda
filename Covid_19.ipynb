```python
# Se importa la librería numpy
import numpy as np

# Vamos a desarrollar un programa de machine learning (básico)
# El siguiente es un paquete de datos a ser procesados:
# La primera columna es: Número de dias
# La segunda columna es: Número de casos
# La tercera columna es : Numero de muertes

data = np.genfromtxt("Covid_19.tsv", delimiter="\t")
#Muestra de los primeros 10 datos
print(data[:10], '\n')

# Número de datos
print(data.shape)
```

```python
# Se divide el array en tres vectores columnas: x, y, z
x = data[:,0]
y = data[:,1]
z = data[:,2]

# Se muestran los valores en x, y, z
print("Vector x[]: \n\n", x, '\n\n')
print("Vector y[]: \n\n", y, '\n\n')
print("Vector z[]: \n\n", z, '\n\n')
```

```python
# Dimensión de los vectores x, y, z
print("Dimension del vector x = ",x.ndim, '\n')
print("Dimension del vector y = ",y.ndim, '\n')
print("Dimension del vector z = ",z.ndim, '\n')

# Elementos contenidos en los vectores x, y, z
print("Contenido del vector x = ",x.shape, '\n')
print("Contenido del vector y = ",y.shape, '\n')
print("Contenido del vector z = ",z.shape)
```

```python
# Investigamos el número de valores nan que contienen los vectores x, y, z

print("Numero de valores nan en el vector x = ",np.sum(np.isnan(x)),'\n')
print("Numero de valores nan en el vector y = ",np.sum(np.isnan(y)),'\n')
print("Numero de valores nan en el vector z = ",np.sum(np.isnan(z)))
```

```python
# Número de elementos en x, y, z, antes de ser comprimidos

print("Contenido del vector x = ",x.shape, '\n')
print("Contenido del vector y = ",y.shape, '\n')
print("Contenido del vector z = ",z.shape, '\n')

# Se eliminan los elementos nan de cada vector
#Se eliminan los elementos en nan en X y esto sirve para eliminar los otras elementos erroneos
z = z[~np.isnan(x)]
y = y[~np.isnan(x)]
x = x[~np.isnan(x)]

# Se eliminan los elemento nan de Y
x = x[~np.isnan(y)]
z = z[~np.isnan(y)]
y = y[~np.isnan(y)]

# Se eliminan los elemento nan de Z
y = y[~np.isnan(z)]
x = x[~np.isnan(z)]
z = z[~np.isnan(z)]


# Se cuenta el número de elementos tanto de x, y, z
print("nuevo contenido del vector x = ",x.shape, '\n')
print("nuevo contenido del vector y = ",y.shape, '\n')
print("nuevo contenido del vector z = ",z.shape, '\n')
```

```python
# Se importa la librería del Sistema Operativo
# Igualmente, la librería utils y numpy
# -----------------------------------------------------------------
import os

# Directorios: chart y data en el directorio de trabajo
# DATA_DIR es el directorio de los datos
# CHART_DIR es el directorio de los gráficos generados
# -----------------------------------------------------------------
from utils import DATA_DIR, CHART_DIR
import numpy as np

# Se eliminan las advertencias por el uso de funciones que
# en el futuro cambiarán
# -----------------------------------------------------------------
np.seterr(all='ignore')

# Se importa la librería scipy y matplotlib
#SciPy contiene módulos para optimización, álgebra lineal, integración, interpolación, funciones especiales, FFT, procesamiento de señales y de imagen, resolución de ODEs y otras tareas para la ciencia e ingeniería.
#SciPy se basa en el objeto de matriz NumPy y es parte del conjunto NumPy, que incluye herramientas como Matplotlib, pandas y SymPy, y un conjunto en expansión de bibliotecas de computación científica. Este conjunto está dirigido al mismo tipo de usuarios que los de aplicaciones como MATLAB, GNU Octave, y Scilab. 
# -----------------------------------------------------------------
import scipy as sp
import matplotlib.pyplot as plt

# Datos de trabajo
# -----------------------------------------------------------------
#La funcion np.genfromtext recibe el nombre y el delimitador(Tabulaciones) separando los datos 
data = np.genfromtxt(os.path.join(DATA_DIR, "Covid_19.tsv"), delimiter="\t")
#En data ,quedan los datos recolectados del archivo Covid_19

# Se establece el tipo de dato donde float64 es un Flotante de 
#doble precisión: bit de signo, exponente de 11 bits, 
#mantisa (se refiere a la diferencia entre un número y su parte entera, es decir, su parte fraccionaria.) de 52 bits 

data = np.array(data, dtype=np.float64)
print(data[:10])
print(data.shape)

# Se definen los colores
# g = green, k = black, b = blue, m = magenta, r = red
# g = verde, k = negro, b = azul, m = magenta, r = rojo
colors = ['g', 'k', 'b', 'm', 'r']

# Se definen los tipos de líneas
# los cuales serán utilizados en las gráficas
linestyles = ['-', '-.', '--', ':', '-']

# Se crea el vector x, correspondiente a la primera columna de data
# Se crea el vector y, correspondiente a la segunda columna de data
# Se crea el vector z, correspondiente a la tercera columna de data
x = data[:, 0]
y = data[:, 1]
z = data[:, 2]

# la función isnan(vector) devuelve un vector en el cual los TRUE
# son valores de tipo nan, y los valores FALSE son valores diferentes
# a nan. Con esta información, este vector permite realizar 
# transformaciones a otros vectores (o al mismo vector), y realizar
# operaciones como sumar el número de posiciones TRUE, con lo
# cual se calcula el total de valores tipo nan
a1 = np.sum(np.isnan(x))
a2 = np.sum(np.isnan(y))
a3 = np.sum(np.isnan(z))
#C es la sumatoria de datos erroneos(nan) que contiene el archivo Covid_19
c=a1+a2+a3
print("Número de entradas incorrectas de x:", a1)
print("Número de entradas incorrectas de y:", a2)
print("Número de entradas incorrectas de z:", a3)
print("Numero de entradas totales incorrectas= ", a1+a2+a3)
# Se eliminan los datos incorrectos
# -----------------------------------------------------------------

# Los valores nan en el vector y deben eliminarse
# Para ello se crea un vector TRUE y FALSE basado en isnan
# Al negar dichos valores (~), los valores que son FALSE se vuelven
# TRUE, y se corresponden con aquellos valores que NO son nan
# Si el vector x, que contiene los valores en el eje x, se afectan
# a partir de dicho valores lógicos, se genera un nuevo vector en
# el que solos se toman aquellos que son TRUE. Por tanto, se crea
# un nuevo vector x, en el cual han desaparecido los correspondientes
# valores de y que son nan

# Esto mismo se aplica, pero sobre el vector y, lo cual hace que tanto
# x como y queden completamente sincronizados: sin valores nan
# Se eliminan los elementos nan de cada vector
#Se eliminan los elementos en nan en x y esto sirve para eliminar los otras elementos erroneos
z = z[~np.isnan(x)]
y = y[~np.isnan(x)]
x = x[~np.isnan(x)]

# Se eliminan los elemento nan de y
x = x[~np.isnan(y)]
z = z[~np.isnan(y)]
y = y[~np.isnan(y)]

# Se eliminan los elemento nan de z
y = y[~np.isnan(z)]
x = x[~np.isnan(z)]
z = z[~np.isnan(z)]

# CON ESTA FUNCIÓN SE DEFINE UN MODELO, EL CUAL CONTIENE 
# el comportamiento de un ajuste con base en un grado polinomial
# elegido
# -----------------------------------------------------------------

#Valores de entrada x,y(Vectores de Covid_19)
#fname Nombre con el cual se va a guardar la grafica
#models modelo implementado 
#None hace referencia a que los valores de entrada que pueden estar vacios

def plot_models(x, y, models, fname, mx=None, ymax=None, xmin=None):
    ''' dibujar datos de entrada '''

    # Crea una nueva figura, o activa una existente.
    # num = identificador, figsize: anchura, altura
    plt.figure(num=None, figsize=(8, 6))
    
    # Borra el espacio de la figura
    plt.clf()
    
    # Un gráfico de dispersión de y frente a x con diferentes tamaños 
    # y colores de marcador (tamaño = 10)
    plt.scatter(x, y, s=10)
    
    # Títulos de la figura
    # Título superior
    plt.title("Covid 19 en Colombia")
    
    # Título en la base
    plt.xlabel("Tiempo en meses")
    
    # Título lateral
    plt.ylabel("Casos positivos")
    
    # Obtiene o establece las ubicaciones de las marcas 
    # actuales y las etiquetas del eje x.
    
    # Los primeros corchetes ([]) se refieren a las marcas en x
    # Los siguientes corchetes ([]) se refieren a las etiquetas
    
    # En el primer corchete se tiene: 1*30 +2*30  + ..., hasta
    # completar el total de puntos en el eje horizontal, según
    # el tamaño del vector x
    
    # Además, se aprovecha para calcular los valores de w, los
    # cuales se agrupan en paquetes de w*30 Esto permite
    # determinar los valores de w desde 1 hasta 100, indicando
    # con ello que se tiene un poco más de 6 meses
    
    # Estos valores se utilizan en el segundo corchete para
    # escribir las etiquetas basadas en estos valores de w
    
    # Por tanto, se escriben etiquetas para w desde 1 hasta
    # 100, lo cual constituye los meses analizadas
    plt.xticks(
        [w * 30 for w in range(100)], 
        ['%i' % w for w in range(100)])
    
#La función plt.xticks realiza la demarcación de cada mes multiplicando la cantidad días aplicándolo en un rango de cien (100). 
#Recibe como valor de entrada cada cuanto debe de hacer la demarcación con un valor de W que funciona como un puntero 
#recorriendo la gráfica contando los puntos, este  proceso de demarcación lo realiza en un rango de cien (100) veces, 
#además recibe una cadena en este caso MESES con una variable %i que funciona como iterador(contador).

    # Aquí se evalúa el tipo de modelo recibido
    # Si no se envía ninguno, no se dibuja ninguna curva de ajuste
    if models:
        
#La función numpy.linspace genera un array NumPy formado por n números equiespaciados entre dos datos. Su sintaxis es:
#numpy.linspace(valor-inicial, valor-final, número de valores)
#El numero de valores se interpreta que el numero de veces que realizara el proceso

        # Si no se define ningún valor para mx (revisar el 
        # código más adelante), el valor de mx será
        # calculado con la función linspace

        # NOTA: linspace devuelve números espaciados uniformemente 
        # durante un intervalo especificado. En este caso, sobre
        # el conjunto de valores x establecido
#En el caso de x[-1],lo realiza hasta el ultimo dato del arreglo x
        if mx is None:
            mx = np.linspace(0, x[-1], 1000)
        
        # La función zip () toma elementos iterables 
        # (puede ser cero o más), los agrega en una tupla y los devuelve
        
        # Aquí se realiza un ciclo .....
        
        
#La zip()función devuelve un objeto zip, que es un iterador de tuplas donde el primer elemento de cada iterador 
#pasado se empareja, y luego el segundo elemento de cada iterador pasado se empareja, etc.     
#Es un emparejamiento por cada elemento segun su orden en el arregllo por ejemplo model sera un vector dentro de otro
#que contiene la lista de los 4 elementos
        for model, style, color in zip(models, linestyles, colors):
            # print "Modelo:",model
            # print "Coeffs:",model.coeffs
            #plt.plot recibe los valores de los vectores,que se van a graficar,ademas tambien recibe las caracteristicas
            #del grafico comoo el tipo de linea ,el tamaño y el color
            #mx se refiere a los puntos de entrada que seran evaluados
            #model(mx),son los puntos evaluados en el modelo producido 
            plt.plot(mx, model(mx), linestyle=style, linewidth=2, c=color)
        #titulo indicando el orden de la ecuacion que se digito 
        #Pone el titulo al lado izquierdo 
        plt.legend(["d=%i" % m.order for m in models], loc="upper left")

#Graficas en matplolib
    plt.autoscale(tight=True)
    plt.ylim(ymin=0)
    if ymax:
        plt.ylim(ymax=ymax)
    #Llamar a esta función sin argumentos (por ejemplo ylim()) es el equivalente en pyplot de llamar 
    #get_ylim a los ejes actuales ,con argumentos deja el limite definido.En caso de llamar ymax o xmin la
    #funcion se encargara de encontrar estos valores
    if xmin:
        plt.xlim(xmin=xmin)
    plt.grid(True, linestyle='-', color='0.75')
    #Guarda la imagen con el respecivo nombre(Fname) al inicio de la funcion
    plt.savefig(fname)

# Primera mirada a los datos
# -----------------------------------------------------------------
plot_models(x, y, None, os.path.join(CHART_DIR, "1400_01_01.png"))
#Grafica los datos con ningun modelo en cuenta

# Crea y dibuja los modelos de datos
# -----------------------------------------------------------------
fp1, res1, rank1, sv1, rcond1 = np.polyfit(x, y, 1, full=True)

#Polyfit es el ajuste  de un polinomio de grado deg a los puntos (x, y) . Devuelve un vector de coeficientes p que minimiza 
#el error al cuadrado en el orden deg , deg-1 ,… 0 .p(x) = p[0] * x**deg + ... + p[deg]

#Ya que full es verdadero devuelve cuatro datos demas (residuos(res1), rango(rank1), valores_individuales(sv1), segundo(rcond))
#Presente solo si full= Verdadero. Los residuos son la suma de los residuos cuadrados del ajuste por mínimos cuadrados, el rango 
#efectivo de la matriz de coeficientes de Vandermonde escalada, sus valores singulares y el valor especificado de rcond . 

#fp1: ndarray, forma (deg + 1,) o (deg + 1, K)
#fp1 es el modelo hayado con coeficientes polinomiales, la mayor potencia primero


#(x,y vectores de entrada y el 1 es el orden del polinomio)
#polyfit permite por ajuste de mínimos cuadrados ver el error presente en los datos
#Los resultados se pueden mejorar reduciendo el grado del polinomio o reemplazando x por x - x .mean ().
#La reduccion de error se realiza con polyd1t ya que permite reducir el grado de polinomio en valores mas simples

#f1 es el modelo fp1 mas preciso,ya que poly1d permite reducir los errores 

print("Parámetros del modelo fp1: %s" % fp1)
#Modelo de ndarray, forma (deg + 1,) o (deg + 1, K)
print("Error del modelo fp1:", res1)
#Los residuos son la suma de los residuos cuadrados del ajuste por mínimos cuadrado.Los residuos son tomados
#como los errores producidos

f1 = sp.poly1d(fp1)
#numpy.poly1d (arr, root, var): esta función ayuda a definir una 
#función polinomial. Facilita la aplicación de "operaciones naturales" en polinomios.
#Pero a simple vista descompone un polinomio en factores de un grado mas bajo 
#En resumen vuelve mas exacto el modelo descoponiendolo en factores mas simples.


#Mismo procedimiento pero con grado 2
fp2, res2, rank2, sv2, rcond2 = np.polyfit(x, y, 2, full=True)
print("Parámetros del modelo fp2: %s" % fp2)
print("Error del modelo fp2:", res2)
f2 = sp.poly1d(fp2)

#Simplificado con grado 3,10 y 100

f3 = sp.poly1d(np.polyfit(x, y, 3))
f10 = sp.poly1d(np.polyfit(x, y, 10))
f100 = sp.poly1d(np.polyfit(x, y, 100))

# Se grafican los modelos
# -----------------------------------------------------------------
plot_models(x, y, [f1], os.path.join(CHART_DIR, "1400_01_02.png"))
plot_models(x, y, [f1, f2], os.path.join(CHART_DIR, "1400_01_03.png"))
plot_models(x, y, [f1, f2, f3, f10, f100], os.path.join(CHART_DIR, "1400_01_04.png"))

#os.path.join()El método en Python une uno o más componentes de ruta de forma inteligente.
#Este método concatena varios componentes de ruta con exactamente un separador de directorio 
#En este caso es el nombre con el cual la grafica va a quedar guardada en la carpeta CHART_DIR con el nombre "1400_01_03.png
#Da como resultado la siguiente direccion CHART_DIR/"1400_01_02.png"

# Ajusta y dibuja un modelo utilizando el conocimiento del punto
# de inflexión
#Un punto de inflexión, en una función matemática, es un punto donde 
#los valores de una función continua en x pasan de un tipo de concavidad a otra
#5.7 especifica el punto donde el crecimiento en la curva empezo a cambiar
# -----------------------------------------------------------------
inflexion = 5.7 * 30
#Toma todos los puntos desde el inicio x hasta el punto inflexion
xa = x[:int(inflexion)]
#Toma todos los puntos desde el inicio y hasta el punto inflexion
ya = y[:int(inflexion)]
#Toma desde el punto inflexion hasta el ultimo valor en x
xb = x[int(inflexion):]
#Toma desde el punto inflexion hasta el ultimo valor en y
yb = y[int(inflexion):]

# Se grafican dos líneas rectas
# -----------------------------------------------------------------
fa = sp.poly1d(np.polyfit(xa, ya, 1))
fb = sp.poly1d(np.polyfit(xb, yb, 1))

#Teniendo los dos modelos se junta en una grafica ,ya que esto es una forma de reducir la complejidad en el 
#modelo matematico ya que el problema se divide en dos no se sobrecarga una funcion haciendo que acapare todos los 
#puntos si no que se le da una zona en especifico 

# Se presenta el modelo basado en el punto de inflexión
# -----------------------------------------------------------------
plot_models(x, y, [fa, fb], os.path.join(CHART_DIR, "1400_01_05.png"))

# Función de error
# -----------------------------------------------------------------
def error(f, x, y):
    return np.sum((f(x) - y) ** 2)

#Se resta el modelo evaluado menos el valor original,para determinar el error o que tanto estan separados de si ,
#se eleva a una potencia de dos por la razon de que algunos errores son negativos ocasionando que el resultado en la sumatoria
#sea igual a cero

# Se imprimen los errores
# -----------------------------------------------------------------
print("Errores para el conjunto completo de datos:")
for f in [f1, f2, f3, f10, f100]:
    print("Error d=%i: %f" % (f.order, error(f, x, y)))
#Ciclo para cada modelo ,con f.order se define el orden del error

print("Errores solamente después del punto de inflexión")
for f in [f1, f2, f3, f10, f100]:
    print("Error d=%i: %f" % (f.order, error(f, xb, yb)))
    
#Se suman los dos errores ya que el modelo se divide en dos por el punto de inflexion    

print("Error de inflexión=%f" % (error(fa, xa, ya) + error(fb, xb, yb)))

# Se extrapola de modo que se proyecten respuestas en el futuro
# -----------------------------------------------------------------
plot_models(
    x, y, [f1, f2, f3, f10, f100],
    os.path.join(CHART_DIR, "1400_01_06.png"),
    mx=np.linspace(0 * 30, 8 * 30, 100),
    ymax=12000, xmin=0 * 30)
#8 *30 Para facilitar el procesamiento define un maximo con un vector de 100 posiciones
#12000 Es una restriccion en pixeles por el alto de la grafica
#Grafica de todos los modelos combinados [f1, f2, f3, f10, f100]

print("Entrenamiento de datos únicamente despúes del punto de inflexión")
fb1 = fb
fb2 = sp.poly1d(np.polyfit(xb, yb, 2))
fb3 = sp.poly1d(np.polyfit(xb, yb, 3))
fb10 = sp.poly1d(np.polyfit(xb, yb, 10))
fb100 = sp.poly1d(np.polyfit(xb, yb, 100))
#Graficas con punto de inflexion con distinto grado de polinomial
#Nota:Entre mayor grado polinomial se asemeja mas al comportamiento de los puntos
#Recordar que entre mayor grado polinomial la complejidad aumenta por que en el modelo
#solo existe por los puntos por lo cual su futuro o prediccion es casi inexistente ya que 
#a la hora de evaluarlo sera casi imposible

print("Errores después del punto de inflexión")
for f in [fb1, fb2, fb3, fb10, fb100]:
    print("Error d=%i: %f" % (f.order, error(f, xb, yb)))
    
#Errores de los modelos con inflexion
# Gráficas después del punto de inflexión
# -----------------------------------------------------------------
plot_models(
    x, y, [fb1, fb2, fb3, fb10, fb100],
    os.path.join(CHART_DIR, "1400_01_07.png"),
    mx=np.linspace(0 * 30, 8*30, 100),
    ymax=12000, xmin=0 * 30)
#8 *30 Para facilitar el procesamiento define un maximo con un vector de 100 posiciones
#10000 Es una restriccion en pixeles por el alto de la grafica
#Grafica de todos los modelos con punto de inflexiom combinados [f1, f2, f3, f10, f100]

# Separa el entrenamiento de los datos de prueba
#Se extrae el 30%,de los datos se mezclan y luego se ordenan respeto al
#punto de inflexion
# -----------------------------------------------------------------
frac = 0.3
#30% en decimal
split_idx = int(frac * len(xb))
#se obtiene el 30% de longitud del vector
shuffled = sp.random.permutation(list(range(len(xb))))
#Se mezclan los indices de los datos
test = sorted(shuffled[:split_idx])
#se toma datos del 30% de la informacion,luego se ordenan 
train = sorted(shuffled[split_idx:])
#se toma datos del 70% de la informacion,luego se ordenan 

#Reduccion de errores con polyfit de los modelos
fbt1 = sp.poly1d(np.polyfit(xb[train], yb[train], 1))
fbt2 = sp.poly1d(np.polyfit(xb[train], yb[train], 2))
print("fbt2(x)= \n%s" % fbt2)
#Modelo cuadratico menos la cantidad de casos 
print("fbt2(x)-6500= \n%s" % (fbt2-6500))
#Modelos con los datos reducidos on orden 3,10,1000
fbt3 = sp.poly1d(np.polyfit(xb[train], yb[train], 3))
fbt10 = sp.poly1d(np.polyfit(xb[train], yb[train], 10))
fbt100 = sp.poly1d(np.polyfit(xb[train], yb[train], 100))
#Errores en los modelos de entrenamiento despues del punto de inflexion
print("Prueba de error para después del punto de inflexión")
for f in [fbt1, fbt2, fbt3, fbt10, fbt100]:
    print("Error d=%i: %f" % (f.order, error(f, xb[test], yb[test])))

#grafica de las funciones
plot_models(
    x, y, [fbt1, fbt2, fbt3, fbt10, fbt100],
    os.path.join(CHART_DIR, "1400_01_08.png"),
    mx=np.linspace(0 * 30, 8 *30, 100),
    ymax=12000, xmin=0 * 30)

from scipy.optimize import fsolve

print(fbt2)
print(fbt2 - 6500)
#6500 es el punto de analisis al que se desea saber el dia en que se va aproximar a ese dato
#Encuentra las raíces de una funcion restando el punto en y para obtener el corte en el eje x, desde el punto de inflexion
#la funcion va recorriendo los puntos en x desde x0 hasta la posicion restada
alcanzado_max = fsolve(fbt2 - 6500, x0=182) / (30)

alcanzado_max2 = fsolve(fbt2 - 6500, x0=182) 
#67 son los dias del año donde comenzo la pandemia
#c desfase de los nan en todo el archivo

dias=int(alcanzado_max2[0])+67+c
print("Dias respecto al año :",dias)

print("\n6530 casos esperados en el mes %f" %alcanzado_max[0])
print("\n6530 casos esperados en el dia %f" %alcanzado_max2[0])

print(fbt2(186))
```

```python

import os

from utils import DATA_DIR, CHART_DIR
import numpy as np

np.seterr(all='ignore')

import scipy as sp
import matplotlib.pyplot as plt

data = np.genfromtxt(os.path.join(DATA_DIR, "Covid_19.tsv"), delimiter="\t")

data = np.array(data, dtype=np.float64)
print(data[:10])
print(data.shape)

colors = ['g', 'k', 'b', 'm', 'r']

linestyles = ['-', '-.', '--', ':', '-']

x = data[:, 0]
y = data[:, 2]
z = data[:, 1]

a1 = np.sum(np.isnan(x))
a2 = np.sum(np.isnan(y))
a3 = np.sum(np.isnan(z))
print("Número de entradas incorrectas de x:", a1)
print("Número de entradas incorrectas de y:", a2)
print("Número de entradas incorrectas de z:", a3)
print("Numero de entradas totales incorrectas= ", a1+a2+a3)

z = z[~np.isnan(x)]
y = y[~np.isnan(x)]
x = x[~np.isnan(x)]

x = x[~np.isnan(y)]
z = z[~np.isnan(y)]
y = y[~np.isnan(y)]

y = y[~np.isnan(z)]
x = x[~np.isnan(z)]
z = z[~np.isnan(z)]

def plot_models(x, y, models, fname, mx=None, ymax=None, xmin=None):
    ''' dibujar datos de entrada '''

    plt.figure(num=None, figsize=(8, 6))
    
    plt.clf()
    
    plt.scatter(x, y, s=10)
    
    plt.title("Covid 19 en Colombia")
    
    plt.xlabel("Tiempo en meses")
    
    plt.ylabel("Muertes")

    plt.xticks(
        [w * 30 for w in range(100)], 
        ['%i' % w for w in range(100)])

    if models:
        

        if mx is None:
            mx = np.linspace(0, x[-1], 1000)
        
        for model, style, color in zip(models, linestyles, colors):
             
            plt.plot(mx, model(mx), linestyle=style, linewidth=2, c=color)
        
        plt.legend(["d=%i" % m.order for m in models], loc="upper left")


    plt.autoscale(tight=True)
    plt.ylim(ymin=0)
    if ymax:
        plt.ylim(ymax=ymax)
   
    if xmin:
        plt.xlim(xmin=xmin)
    plt.grid(True, linestyle='-', color='0.75')
    plt.savefig(fname)


plot_models(x, y, None, os.path.join(CHART_DIR, "2400_01_01.png"))

fp1, res1, rank1, sv1, rcond1 = np.polyfit(x, y, 1, full=True)

print("Parámetros del modelo fp1: %s" % fp1)

print("Error del modelo fp1:", res1)

f1 = sp.poly1d(fp1)

fp2, res2, rank2, sv2, rcond2 = np.polyfit(x, y, 2, full=True)
print("Parámetros del modelo fp2: %s" % fp2)
print("Error del modelo fp2:", res2)
f2 = sp.poly1d(fp2)

f3 = sp.poly1d(np.polyfit(x, y, 3))
f10 = sp.poly1d(np.polyfit(x, y, 10))
f100 = sp.poly1d(np.polyfit(x, y, 100))

plot_models(x, y, [f1], os.path.join(CHART_DIR, "2400_01_02.png"))
plot_models(x, y, [f1, f2], os.path.join(CHART_DIR, "2400_01_03.png"))
plot_models(x, y, [f1, f2, f3, f10, f100], os.path.join(CHART_DIR, "2400_01_04.png"))

inflexion = 5.5 * 30

xa = x[:int(inflexion)]

ya = y[:int(inflexion)]

xb = x[int(inflexion):]

yb = y[int(inflexion):]

fa = sp.poly1d(np.polyfit(xa, ya, 1))

fb = sp.poly1d(np.polyfit(xb, yb, 1))

plot_models(x, y, [fa, fb], os.path.join(CHART_DIR, "2400_01_05.png"))

def error(f, x, y):
    return np.sum((f(x) - y) ** 2)

print("Errores para el conjunto completo de datos:")
for f in [f1, f2, f3, f10, f100]:
    print("Error d=%i: %f" % (f.order, error(f, x, y)))

print("Errores solamente después del punto de inflexión")
for f in [f1, f2, f3, f10, f100]:
    print("Error d=%i: %f" % (f.order, error(f, xb, yb)))
    
print("Error de inflexión=%f" % (error(fa, xa, ya) + error(fb, xb, yb)))


plot_models(
    x, y, [f1, f2, f3, f10, f100],
    os.path.join(CHART_DIR, "2400_01_06.png"),
    mx=np.linspace(0 * 30, 8 * 30, 100),
    ymax=450, xmin=0 * 30)


print("Entrenamiento de datos únicamente despúes del punto de inflexión")
fb1 = fb
fb2 = sp.poly1d(np.polyfit(xb, yb, 2))
fb3 = sp.poly1d(np.polyfit(xb, yb, 3))
fb10 = sp.poly1d(np.polyfit(xb, yb, 10))
fb100 = sp.poly1d(np.polyfit(xb, yb, 100))

print("Errores después del punto de inflexión")
for f in [fb1, fb2, fb3, fb10, fb100]:
    print("Error d=%i: %f" % (f.order, error(f, xb, yb)))
    
plot_models(
    x, y, [fb1, fb2, fb3, fb10, fb100],
    os.path.join(CHART_DIR, "2400_01_07.png"),
    mx=np.linspace(0 * 30, 8*30, 100),
    ymax=450, xmin=0 * 30)

frac = 0.3
split_idx = int(frac * len(xb))
shuffled = sp.random.permutation(list(range(len(xb))))
test = sorted(shuffled[:split_idx])
train = sorted(shuffled[split_idx:])
fbt1 = sp.poly1d(np.polyfit(xb[train], yb[train], 1))
fbt2 = sp.poly1d(np.polyfit(xb[train], yb[train], 2))
print("fbt2(x)= \n%s" % fbt2)
print("fbt2(x)-100,000= \n%s" % (fbt2-185))
fbt3 = sp.poly1d(np.polyfit(xb[train], yb[train], 3))
fbt10 = sp.poly1d(np.polyfit(xb[train], yb[train], 10))
fbt100 = sp.poly1d(np.polyfit(xb[train], yb[train], 100))

print("Prueba de error para después del punto de inflexión")
for f in [fbt1, fbt2, fbt3, fbt10, fbt100]:
    print("Error d=%i: %f" % (f.order, error(f, xb[test], yb[test])))

plot_models(
    x, y, [fbt1, fbt2, fbt3, fbt10, fbt100],
    os.path.join(CHART_DIR, "2400_01_08.png"),
    mx=np.linspace(0 * 30, 8 *30, 100),
    ymax=450, xmin=0 * 30)

from scipy.optimize import fsolve
print(fbt2)
print(fbt2 - 185)
alcanzado_max = fsolve(fbt2 - 185, x0=171) / (30)
print("\n185 casos esperados en el mes %f" % 
      alcanzado_max[0])

print(fbt2(186))
```
